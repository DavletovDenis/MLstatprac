### Merge the table ```dfa``` that contains order date and order id with ```df``` and plot daily revenue 

````python
dfa = pd.DataFrame({"date": np.random.choice(pd.date_range(start="20170101", end="20170201"), size=1800),
                   "order_id": np.arange(1800)})
````

````python
dfa.head(3)
````

````python
df1 = pd.merge(dfa, df, on='order_id')
df1.head(10)
````

````python
date = df1['date'].unique()
````

````python
date = sorted(date)
````

````python
df1[df1['date'] == date[1]]['item_price']
````

````python
revenues = []
for i in date:
    revenues.append(sum(list(df1[df1['date'] == i]['item_price'])))
````

````python
plt.figure(figsize=(15,8))
plt.grid()
plt.plot(date, revenues)
plt.xlabel('Day')
plt.ylabel('Revenue')
````

### Create a histogram of the top 5 items bought

````python
top5 = dict()
````

````python
a = []
````

````python
for i in unique:
    a.append(sum(list(df[df['item_name'] == i]['quantity'])))
````

````python
a = sorted(a)[::-1]
a = a[:5]
````

````python
for i in a:
    for j in unique:
        if sum(list(df[df['item_name'] == j]['quantity'])) == i:
            top5[j] = i
````

````python
top5
````

````python
plt.figure(figsize=(10,8))
plt.grid()
plt.bar(range(len(top5)), list(top5.values()), align='center')
plt.xticks(range(len(top5)), list(top5.keys()))
````
