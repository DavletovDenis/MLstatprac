```python
def e_step(y, mu1_cur, mu2_cur, pi1_cur):
    pi2_cur = 1 - pi1_cur
    r_i1 = binom.pmf(y, 10, mu1_cur) * pi1_cur / (binom.pmf(y, 10, mu1_cur) * pi1_cur  + binom.pmf(y, 10, mu2_cur) * pi2_cur)
    r_i2 = binom.pmf(y, 10, mu2_cur) * pi2_cur / (binom.pmf(y, 10, mu1_cur) * pi1_cur  + binom.pmf(y, 10, mu2_cur) * pi2_cur)
    return r_i1, r_i2
```

```python
def m_step(y, r1, r2):
    pi1_new = sum(r1) / len(y)
    pi2_new = 1 - pi1_new
    mu1_new = sum(y * r1) / (10 * sum(r1))
    mu2_new = sum(y * r2) / (10 * sum(r2))
    return pi1_new, pi2_new, mu1_new, mu2_new
```

```python
def em(y):
    mu1 = np.random.random()
    mu2 = np.random.random()
    pi1 = np.random.random()
    pi2 = 1 - pi1
    mu1_list = [mu1]
    mu2_list = [mu2]
    pi1_list = [pi1]
    pi2_list = [pi2]
    while True:
        r1, r2 = e_step(y, mu1, mu2, pi1)
        pi1_new, pi2_new, mu1_new, mu2_new = m_step(y, r1, r2)
       # df.append({'pi1' : pi1_new, 'pi2' : pi2_new, 'mu1' : mu1_new, 'mu2' : mu2_new}, ignore_index = True)
        mu1_list.append(mu1_new)
        mu2_list.append(mu2_new)
        pi1_list.append(pi1_new)
        pi2_list.append(pi2_new)
        if np.sqrt((mu1 - mu1_new)**2 + (mu2 - mu2_new)**2 + 2 * (pi1 - pi1_new)**2) < 0.00001:
            break
        mu1, mu2, pi1 = mu1_new, mu2_new, pi1_new
    df = pd.DataFrame({'pi1' : pi1_list, 'pi2' : pi2_list, 'mu1' : mu1_list, 'mu2' : mu2_list})
    print(len(df), 'итераций')
    return df
```
